```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('fence.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap = "gray")
plt.title("input")
plt.show()
```

```python
img_ft = np.fft.fft2(img)
img_ftshift = np.fft.fftshift(img_ft)

plt.imshow(np.log(1 + np.abs(img_ftshift)))
plt.title("img ftshift")
plt.colorbar()
plt.show()
```

```python
N = 20
h = np.ones((N, N)) / (N*N)

#print(h)

h = np.pad(h, [(0, img.shape[0]-h.shape[0]), (0, img.shape[1]-h.shape[1])], mode='constant')

#print(h)

h_ft = np.fft.fft2(h)
h_ftshift = np.fft.fftshift(h_ft)

plt.imshow(np.log(1 + np.abs(h_ftshift)))
plt.title("h ftshift")
plt.colorbar()
plt.show()
```

```python
g_ftshift = img_ftshift * h_ftshift

plt.imshow(np.log(1 + np.abs(g_ftshift)))
plt.title("g ftshift")
plt.colorbar()
plt.show()
```

```python
g_ft = np.fft.ifftshift(g_ftshift)

imgOut = np.real(np.fft.ifft2(g_ft))

plt.imshow(imgOut, cmap="gray")
plt.title("imgOut")
plt.colorbar()
plt.show()
```
