```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

BGR_img = cv2.imread('snow.png') # reading the image as bgr
print(BGR_img.shape)

RGB_img = cv2.cvtColor(BGR_img, cv2.COLOR_BGR2RGB)

print("RGB version of the image")
plt.imshow(RGB_img)
plt.show()
```

```python
HSV_img = cv2.cvtColor(BGR_img, cv2.COLOR_BGR2HSV) # transforming it into HSV form

print("HSV version of the image")
plt.imshow(HSV_img)
plt.show()
```

```python
plt.imshow(HSV_img[:,:,2], cmap='gray') #showing its V channel
plt.title(f'V Channel') 
plt.show()
```

```python
V_hist = cv2.calcHist(HSV_img[:,:,2],[0], None, [256], [0, 255]) # creating the V channel histogram 
plt.plot(V_hist, color='b', label="HUE") # defining its characteristics
plt.title('V channel histogram')
plt.xlim(30,60)
plt.ylim(0,150)
plt.show() 
```

```python
HSV_img[:,:,2] = cv2.equalizeHist(HSV_img[:, :, 2])
image = cv2.cvtColor(HSV_img, cv2.COLOR_HSV2RGB)
plt.title(f'Image after histeq') 
plt.imshow(image)
plt.show()
```

```python
plt.imshow(HSV_img[:,:,2], cmap='gray') #showing its V channel
plt.title(f'V Channel after histeq') 
plt.show()
```

```python
V_hist = cv2.calcHist(HSV_img[:,:,2],[0], None, [256], [0, 255]) # creating the V channel histogram 
plt.plot(V_hist, color='b', label="HUE") # defining its characteristics
plt.title('V channel after histeq')
plt.xlim(25,100)
plt.ylim(0,150)
plt.show() 
```
