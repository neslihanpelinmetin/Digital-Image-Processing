```python
import cv2
from matplotlib import pyplot as plt

!wget "https://upload.wikimedia.org/wikipedia/commons/thumb/0/0f/1665_Girl_with_a_Pearl_Earring.jpg/422px-1665_Girl_with_a_Pearl_Earring.jpg"
```

```python
BGR_img = cv2.imread('422px-1665_Girl_with_a_Pearl_Earring.jpg')

RGB_img = cv2.cvtColor(BGR_img, cv2.COLOR_BGR2RGB)
print("Original image\n")
plt.imshow(RGB_img)
plt.show()

(row, col) = BGR_img.shape[0:2]
  
# Take the average of pixel values of the BGR Channels
# to convert the colored image to grayscale image
for i in range(row):
    for j in range(col):
        # Find the average of the BGR pixel values
        BGR_img[i, j] = sum(BGR_img[i, j]) * 0.33
  
print("Gray image\n")
plt.imshow(BGR_img)
plt.show()
```

```python
import numpy as np

kernel = np.array([[1, 0, -1],
                   [2, 0, -2],
                   [1, 0, -1]])

print("Gray image with Filter 1 \n")
dst1 = cv2.filter2D(BGR_img, -1, kernel)
plt.imshow(dst1)
plt.show()

dst3 = cv2.filter2D(BGR_img, -1, kernel)
```

```python
kernel2 = np.array([[1, 2, 1],
                   [0, 0, 0],
                   [-1, -2, -1]])

print("Gray image with Filter 2 \n")
dst2 = cv2.filter2D(BGR_img, -1, kernel2)
plt.imshow(dst2)
plt.show()


print("Gray image with both of the Filters \n")
dst3 = cv2.filter2D(dst3, -1, kernel2)
plt.imshow(dst3)
plt.show()
```

```python
print("Gray image with both of the Filters converted to a colormap\n")
im_color0 = cv2.applyColorMap(dst3, cv2.COLORMAP_JET)
plt.imshow(im_color0)
plt.show()

im_color0 = cv2.applyColorMap(dst3, cv2.COLORMAP_RAINBOW)
plt.imshow(im_color0)
plt.show()
```
