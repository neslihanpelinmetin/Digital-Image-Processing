```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('moon.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap = "gray")
plt.title("input")
plt.show()
```

```python
img_ft = np.fft.fft2(img)
img_ftshift = np.fft.fftshift(img_ft)

plt.imshow(np.log(1 + np.abs(img_ftshift)))
plt.title("img ftshift")
plt.colorbar()
plt.show()
```

```python
height, width = img_ftshift.shape

R = 20
H = np.zeros((height, width))

for r in range(height):
    for c in range(width):
        dist = np.linalg.norm([r-height/2, c-width/2])
        if dist <= R:
            H[r][c] = 1

plt.imshow(H)
plt.title("H")
plt.colorbar()
plt.show()
```

```python
g_ftshift = img_ftshift * H

plt.imshow(np.log(1 + np.abs(g_ftshift)))
plt.title("g ftshift")
plt.colorbar()
plt.show()
```

```python
g_ft = np.fft.ifftshift(g_ftshift)

imgOut = np.real(np.fft.ifft2(g_ft))

plt.imshow(imgOut, cmap="gray")
plt.title("imgOut")
plt.colorbar()
plt.show()
```
