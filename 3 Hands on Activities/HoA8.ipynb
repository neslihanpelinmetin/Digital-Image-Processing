```python
import cv2
from matplotlib import pyplot as plt

starry = cv2.imread('Temple+of+Efes+Artemis++reconstruction,+circa+350-300+BC+....jpg')
starry = cv2.cvtColor(starry, cv2.COLOR_BGR2GRAY)

plt.imshow(starry, cmap = 'gray')
plt.show()
```

```python
import numpy as np

F = np.fft.fft2(starry)

plt.imshow(np.log(1 + np.abs(F)), cmap = 'gray')
plt.title("img ft")
plt.colorbar()
plt.show()
```

```python
Fshift = np.fft.fftshift(F)

plt.imshow(np.log(1 + np.abs(Fshift)), cmap = 'gray')
plt.title("img ftshift")
plt.colorbar()
plt.show()
```

```python
fft_img_unshifted = np.fft.ifftshift(Fshift)

img_back = np.fft.ifft2(fft_img_unshifted)

max_diff = np.max(np.abs(starry.astype(float) - img_back.real))
print("Maximum absolute difference between original and reconstructed image: ", max_diff)

plt.imshow(starry, cmap='gray')
plt.title('Original Image')
plt.show()

plt.imshow(img_back.real, cmap='gray')
plt.title('Reconstructed Image')

plt.show()
```
