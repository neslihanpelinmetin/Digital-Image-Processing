```python
import cv2
from matplotlib import pyplot as plt

im = cv2.imread('aglayan-kadin-picasso-hikayesi.jpg?v=1655806799')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(im, cmap = 'gray')
plt.show()
```

```python
cannyIm = cv2.Canny(im, 100, 200)
plt.imshow(cannyIm, cmap = 'gray')
plt.show()
```

```python
cannyIm2 = cv2.Canny(im, 30, 200)
plt.imshow(cannyIm2, cmap = 'gray')
plt.show()
```

```python
cannyIm3 = cv2.Canny(im, 300, 200)
plt.imshow(cannyIm3, cmap = 'gray')
plt.show()
```
